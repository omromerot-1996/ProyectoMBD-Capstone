```python
from google.colab import files
uploaded = files.upload()

!pip -q install tbats
!pip -q install pmdarima
!pip -q install statsmodels scikit-learn tensorflow openpyxl


import numpy as np
import pandas as pd
import unicodedata
import warnings
warnings.filterwarnings("ignore")

from statsmodels.tsa.stattools import adfuller, kpss


def norm_col(c):
    c = str(c).strip().lower()
    c = unicodedata.normalize("NFKD", c).encode("ascii", "ignore").decode("utf-8")
    c = c.replace(" ", "_").replace("-", "_")
    while "__" in c:
        c = c.replace("__", "_")
    return c


MUESTRA = "Muestra.xlsx"
F2014_2024 = "mdi_homicidios_intencionales_pm_2014_2024.xlsx"
F2025 = "mdi_homicidiosintencionalse_pm_2025_enero_octubre.xlsx"


df_muestra = pd.read_excel(MUESTRA, engine="openpyxl")
schema_cols = [norm_col(c) for c in df_muestra.columns]
schema_set = set(schema_cols)


def read_mdi(path):
    xls = pd.ExcelFile(path, engine="openpyxl")
    sheet = "1" if "1" in xls.sheet_names else xls.sheet_names[-1]
    d = pd.read_excel(path, sheet_name=sheet, engine="openpyxl")
    d.columns = [norm_col(c) for c in d.columns]
    return d


def align_to_schema(df):
    keep = [c for c in df.columns if c in schema_set]
    out = df[keep].copy()
    for c in schema_cols:
        if c not in out.columns:
            out[c] = pd.NA
    return out[schema_cols]


df1 = align_to_schema(read_mdi(F2014_2024))
df2 = align_to_schema(read_mdi(F2025))
df = pd.concat([df1, df2], ignore_index=True)


df["fecha_infraccion"] = pd.to_datetime(df["fecha_infraccion"], errors="coerce")
df = df.dropna(subset=["fecha_infraccion"]).drop_duplicates().copy()

df["provincia"] = df["provincia"].astype(str).str.upper().str.strip()
df["canton"] = df["canton"].astype(str).str.upper().str.strip()

start = pd.Timestamp("2021-01-01")
end = pd.Timestamp("2025-12-31")
df = df[(df["fecha_infraccion"] >= start) & (df["fecha_infraccion"] <= end)].copy()


def weekly_series(d):
    ts = (
        d.set_index("fecha_infraccion")
         .resample("W-SUN")
         .size()
         .rename("homicidios")
    )
    return ts.asfreq("W-SUN", fill_value=0).astype(float)


ts_total = weekly_series(df)
ts_guayas = weekly_series(df[df["provincia"] == "GUAYAS"])

if "GUAYAQUIL" in df["canton"].unique():
    ts_gye = weekly_series(df[df["canton"] == "GUAYAQUIL"])
elif "GUAYQUIL" in df["canton"].unique():
    ts_gye = weekly_series(df[df["canton"] == "GUAYQUIL"])
else:
    ts_gye = None


def run_stationarity_tests(ts, label):
    ts = ts.dropna().astype(float)

    print(f"\n================ {label} | Serie semanal =================")
    print(f"Rango: {ts.index.min().date()} a {ts.index.max().date()} | N={len(ts)}")

    adf_level = adfuller(ts, autolag="AIC")
    print("\n=== ADF | Niveles ===")
    print(f"ADF: {adf_level[0]:.4f} | p-value: {adf_level[1]:.4f}")
    print("CrÃ­ticos:", {k: round(v, 4) for k, v in adf_level[4].items()})

    kpss_level = kpss(ts, regression="c", nlags="auto")
    print("\n=== KPSS | Niveles ===")
    print(f"KPSS: {kpss_level[0]:.4f} | p-value: {kpss_level[1]:.4f}")
    print("CrÃ­ticos:", {k: round(v, 4) for k, v in kpss_level[3].items()})

    ts_diff = ts.diff().dropna()

    adf_diff = adfuller(ts_diff, autolag="AIC")
    print("\n=== ADF | 1ra diferencia ===")
    print(f"ADF: {adf_diff[0]:.4f} | p-value: {adf_diff[1]:.4f}")
    print("CrÃ­ticos:", {k: round(v, 4) for k, v in adf_diff[4].items()})

    kpss_diff = kpss(ts_diff, regression="c", nlags="auto")
    print("\n=== KPSS | 1ra diferencia ===")
    print(f"KPSS: {kpss_diff[0]:.4f} | p-value: {kpss_diff[1]:.4f}")
    print("CrÃ­ticos:", {k: round(v, 4) for k, v in kpss_diff[3].items()})


run_stationarity_tests(ts_total, "TOTAL PAÃS")
run_stationarity_tests(ts_guayas, "PROVINCIA: GUAYAS")

if ts_gye is not None:
    run_stationarity_tests(ts_gye, "CANTÃ“N: GUAYAQUIL")
else:
    print("\n[AVISO] No se encontrÃ³ GUAYAQUIL/GUAYQUIL en 'canton' para construir serie cantonal.")


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import unicodedata
import warnings
warnings.filterwarnings("ignore")


def norm_col(c):
    c = str(c).strip().lower()
    c = unicodedata.normalize("NFKD", c).encode("ascii", "ignore").decode("utf-8")
    c = c.replace(" ", "_").replace("-", "_")
    while "__" in c:
        c = c.replace("__", "_")
    return c


F2014_2024 = "mdi_homicidios_intencionales_pm_2014_2024.xlsx"
F2025 = "mdi_homicidiosintencionalse_pm_2025_enero_octubre.xlsx"


def read_mdi(path):
    xls = pd.ExcelFile(path)
    sheet = "1" if "1" in xls.sheet_names else xls.sheet_names[-1]
    d = pd.read_excel(path, sheet_name=sheet)
    d.columns = [norm_col(c) for c in d.columns]
    return d


df1 = read_mdi(F2014_2024)
df2 = read_mdi(F2025)
df = pd.concat([df1, df2], ignore_index=True).drop_duplicates()

df["fecha_infraccion"] = pd.to_datetime(df["fecha_infraccion"], errors="coerce")
df = df.dropna(subset=["fecha_infraccion"]).sort_values("fecha_infraccion").copy()

ts_weekly = (
    df.set_index("fecha_infraccion")
      .resample("W-SUN")
      .size()
      .rename("homicidios_semanales")
      .asfreq("W-SUN", fill_value=0)
      .astype(float)
)

dfw = ts_weekly.to_frame()
dfw["anio"] = dfw.index.year
dfw["semana_anio"] = dfw.index.isocalendar().week.astype(int)
dfw["t"] = np.arange(1, len(dfw) + 1)


plt.figure(figsize=(10, 5))
plt.plot(dfw.index, dfw["homicidios_semanales"])
plt.title("EvoluciÃ³n semanal de homicidios en Ecuador (2014â€“2025)")
plt.xlabel("Fecha")
plt.ylabel("Homicidios semanales")
plt.tight_layout()
plt.show()


years = sorted(dfw["anio"].unique())
data_by_year = [dfw.loc[dfw["anio"] == y, "homicidios_semanales"].values for y in years]

plt.figure(figsize=(12, 5))
plt.boxplot(data_by_year, labels=years, showfliers=True)
plt.title("DistribuciÃ³n anual de homicidios semanales")
plt.xlabel("AÃ±o")
plt.ylabel("Homicidios semanales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


season_week = (
    dfw.groupby("semana_anio")["homicidios_semanales"]
       .mean()
       .reset_index()
       .sort_values("semana_anio")
)

plt.figure(figsize=(10, 5))
plt.plot(season_week["semana_anio"], season_week["homicidios_semanales"], marker="o")
plt.title("PatrÃ³n estacional promedio de homicidios (semana del aÃ±o)")
plt.xlabel("Semana del aÃ±o")
plt.ylabel("Promedio de homicidios semanales")
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 5))
plt.scatter(dfw["t"], dfw["homicidios_semanales"])
plt.title("DispersiÃ³n temporal de homicidios semanales")
plt.xlabel("Tiempo (Ã­ndice)")
plt.ylabel("Homicidios semanales")
plt.tight_layout()
plt.show()


corr_df = dfw[["homicidios_semanales", "t", "semana_anio"]].copy()
corr = corr_df.corr(numeric_only=True)

plt.figure(figsize=(6, 5))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.index)), corr.index)
plt.title("Matriz de correlaciÃ³n")
plt.tight_layout()
plt.show()


!pip -q install statsmodels tensorflow openpyxl pmdarima


import pandas as pd
import numpy as np
import unicodedata
import warnings
warnings.filterwarnings("ignore")

from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

import pmdarima as pm
from IPython.display import display


np.random.seed(42)
tf.random.set_seed(42)
tf.get_logger().setLevel("ERROR")


def norm_col(c):
    c = str(c).strip().lower()
    c = unicodedata.normalize("NFKD", c).encode("ascii", "ignore").decode("utf-8")
    c = c.replace(" ", "_").replace("-", "_")
    while "__" in c:
        c = c.replace("__", "_")
    return c


MUESTRA = "Muestra.xlsx"
F2014_2024 = "mdi_homicidios_intencionales_pm_2014_2024.xlsx"
F2025 = "mdi_homicidiosintencionalse_pm_2025_enero_octubre.xlsx"


df_muestra = pd.read_excel(MUESTRA, engine="openpyxl")
schema_cols = [norm_col(c) for c in df_muestra.columns]
need = ["fecha_infraccion", "provincia", "canton"]

for r in need:
    if r not in schema_cols:
        raise ValueError(f"âŒ Muestra.xlsx no contiene la columna requerida: {r}")

schema_set = set(schema_cols)


def read_mdi(path):
    xls = pd.ExcelFile(path, engine="openpyxl")
    sheet = "1" if "1" in xls.sheet_names else xls.sheet_names[-1]
    d = pd.read_excel(path, sheet_name=sheet, engine="openpyxl")
    d.columns = [norm_col(c) for c in d.columns]
    return d


def align_to_schema(df):
    keep = [c for c in df.columns if c in schema_set]
    out = df[keep].copy()
    for c in schema_cols:
        if c not in out.columns:
            out[c] = pd.NA
    return out[schema_cols]


df1 = align_to_schema(read_mdi(F2014_2024))
df2 = align_to_schema(read_mdi(F2025))
df = pd.concat([df1, df2], ignore_index=True)

df["fecha_infraccion"] = pd.to_datetime(df["fecha_infraccion"], errors="coerce")
df = df.dropna(subset=["fecha_infraccion"]).copy()
df["provincia"] = df["provincia"].astype(str).str.upper().str.strip()
df["canton"] = df["canton"].astype(str).str.upper().str.strip()

start = pd.Timestamp("2021-01-01")
end = pd.Timestamp("2025-12-31")
df = df[(df["fecha_infraccion"] >= start) & (df["fecha_infraccion"] <= end)].copy()

if df.empty:
    raise ValueError("âŒ Con el filtro 2021-2025 no quedÃ³ data. Revisa fechas en tus archivos.")


def weekly_series(d):
    ts = (
        d.set_index("fecha_infraccion")
         .resample("W-SUN")
         .size()
         .rename("homicidios")
    )
    return ts.asfreq("W-SUN", fill_value=0)


ts_total = weekly_series(df)
ts_guayas = weekly_series(df[df["provincia"] == "GUAYAS"])
ts_guayaquil = weekly_series(df[df["canton"] == "GUAYAQUIL"])


def metrics(y_true, y_pred):
    y_true = np.array(y_true, dtype=float)
    y_pred = np.array(y_pred, dtype=float)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    denom = np.where(y_true == 0, np.nan, y_true)
    mape = np.nanmean(np.abs((y_true - y_pred) / denom)) * 100
    return rmse, mae, mape


def split_fixed(ts, train_weeks=26, test_weeks=26):
    ts = ts.dropna()
    need = train_weeks + test_weeks
    if len(ts) < need:
        raise ValueError(f"âŒ Serie demasiado corta ({len(ts)} semanas). Necesitas al menos {need}.")
    ts_cut = ts.iloc[-need:]
    train = ts_cut.iloc[:train_weeks]
    test = ts_cut.iloc[train_weeks:]
    return train, test


def run_arima(train, test):
    z_train = np.log1p(train.values.astype(float))
    m = pm.auto_arima(
        z_train,
        seasonal=False,
        stepwise=True,
        suppress_warnings=True,
        error_action="ignore",
        max_p=2, max_q=2, max_d=2,
        information_criterion="aic",
        n_jobs=1
    )
    z_fc = m.predict(n_periods=len(test))
    fc = np.expm1(z_fc)
    fc = np.clip(fc, 0, None)
    return (*metrics(test.values, fc), str(m.order))


def run_holt_winters(train, test):
    z_train = np.log1p(train.values.astype(float))
    candidates = [
        dict(trend="add", damped_trend=False),
        dict(trend="add", damped_trend=True),
        dict(trend=None, damped_trend=False),
    ]
    best = None
    best_sse = np.inf
    best_cfg = None

    for cfg in candidates:
        try:
            m = ExponentialSmoothing(
                z_train,
                trend=cfg["trend"],
                seasonal=None,
                damped_trend=cfg["damped_trend"]
            ).fit(optimized=True)
            if m.sse < best_sse:
                best_sse = m.sse
                best = m
                best_cfg = cfg
        except:
            continue

    if best is None:
        m = ExponentialSmoothing(z_train, trend="add", seasonal=None).fit()
        best = m
        best_cfg = dict(trend="add", damped_trend=False)

    z_fc = best.forecast(len(test))
    fc = np.expm1(z_fc)
    fc = np.clip(fc, 0, None)
    return (*metrics(test.values, fc), str(best_cfg))


def run_lstm(train, test, lookback=8):
    y = train.values.astype(float)
    y_log = np.log1p(y)

    scaler = MinMaxScaler()
    train_scaled = scaler.fit_transform(y_log.reshape(-1, 1))

    lb = min(lookback, max(3, len(train_scaled) // 2))

    X_train, y_train = [], []
    for i in range(lb, len(train_scaled)):
        X_train.append(train_scaled[i - lb:i])
        y_train.append(train_scaled[i])
    X_train = np.array(X_train)
    y_train = np.array(y_train)

    model = Sequential([
        LSTM(16, input_shape=(lb, 1)),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")

    es = EarlyStopping(monitor="loss", patience=6, restore_best_weights=True)
    model.fit(X_train, y_train, epochs=80, batch_size=8, verbose=0, callbacks=[es])

    history = train_scaled.copy()
    preds_scaled = []
    for _ in range(len(test)):
        x = history[-lb:].reshape(1, lb, 1)
        p = model.predict(x, verbose=0)[0, 0]
        preds_scaled.append(p)
        history = np.vstack([history, [[p]]])

    preds_log = scaler.inverse_transform(np.array(preds_scaled).reshape(-1, 1)).flatten()
    preds = np.expm1(preds_log)
    preds = np.clip(preds, 0, None)

    return (*metrics(test.values, preds), f"lookback={lb}, units=16, epochs<=80")


def evaluate(ts, label, train_weeks=26, test_weeks=26):
    train, test = split_fixed(ts, train_weeks=train_weeks, test_weeks=test_weeks)

    rows = []
    rmse, mae, mape, params = run_arima(train, test)
    rows.append(["ARIMA", rmse, mae, mape, params])

    rmse, mae, mape, params = run_holt_winters(train, test)
    rows.append(["Holt-Winters", rmse, mae, mape, params])

    rmse, mae, mape, params = run_lstm(train, test)
    rows.append(["LSTM", rmse, mae, mape, params])

    out = (
        pd.DataFrame(rows, columns=["Modelo", "RMSE", "MAE", "MAPE", "Params"])
          .sort_values("RMSE")
          .reset_index(drop=True)
    )

    print(f"\nðŸ“ {label}")
    print(f"Train: {train.index.min().date()} a {train.index.max().date()}  ({len(train)} semanas)")
    print(f"Test : {test.index.min().date()} a {test.index.max().date()}  ({len(test)} semanas)")
    display(out)
    return out


res_total = evaluate(ts_total, "TOTAL PAÃS", 26, 26)
res_guayas = evaluate(ts_guayas, "PROVINCIA: GUAYAS", 26, 26)
res_guayaquil = evaluate(ts_guayaquil, "CANTÃ“N: GUAYAQUIL", 26, 26)

print("\nâœ… Hecho. Si quieres, agrego grÃ¡ficos + predicciÃ³n futura del mejor modelo por nivel.")


!pip -q install statsmodels tensorflow openpyxl pmdarima


import pandas as pd
import numpy as np
import unicodedata
import warnings
warnings.filterwarnings("ignore")

from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

import pmdarima as pm
from IPython.display import display


np.random.seed(42)
tf.random.set_seed(42)
tf.get_logger().setLevel("ERROR")


def norm_col(c):
    c = str(c).strip().lower()
    c = unicodedata.normalize("NFKD", c).encode("ascii", "ignore").decode("utf-8")
    c = c.replace(" ", "_").replace("-", "_")
    while "__" in c:
        c = c.replace("__", "_")
    return c


MUESTRA = "Muestra.xlsx"
F2014_2024 = "mdi_homicidios_intencionales_pm_2014_2024.xlsx"
F2025 = "mdi_homicidiosintencionalse_pm_2025_enero_octubre.xlsx"


df_muestra = pd.read_excel(MUESTRA, engine="openpyxl")
schema_cols = [norm_col(c) for c in df_muestra.columns]
need = ["fecha_infraccion", "provincia", "canton"]

for r in need:
    if r not in schema_cols:
        raise ValueError(f"âŒ Muestra.xlsx no contiene: {r}")

schema_set = set(schema_cols)


def read_mdi(path):
    xls = pd.ExcelFile(path, engine="openpyxl")
    sheet = "1" if "1" in xls.sheet_names else xls.sheet_names[-1]
    d = pd.read_excel(path, sheet_name=sheet, engine="openpyxl")
    d.columns = [norm_col(c) for c in d.columns]
    return d


def align_to_schema(df):
    keep = [c for c in df.columns if c in schema_set]
    out = df[keep].copy()
    for c in schema_cols:
        if c not in out.columns:
            out[c] = pd.NA
    return out[schema_cols]


df1 = align_to_schema(read_mdi(F2014_2024))
df2 = align_to_schema(read_mdi(F2025))
df = pd.concat([df1, df2], ignore_index=True)

df["fecha_infraccion"] = pd.to_datetime(df["fecha_infraccion"], errors="coerce")
df = df.dropna(subset=["fecha_infraccion"]).copy()
df["provincia"] = df["provincia"].astype(str).str.upper().str.strip()
df["canton"] = df["canton"].astype(str).str.upper().str.strip()

start = pd.Timestamp("2021-01-01")
end = pd.Timestamp("2025-12-31")
df = df[(df["fecha_infraccion"] >= start) & (df["fecha_infraccion"] <= end)].copy()

if df.empty:
    raise ValueError("âŒ No quedÃ³ data en 2021-2025. Revisa fechas.")


def weekly_series(d):
    ts = (
        d.set_index("fecha_infraccion")
         .resample("W-SUN")
         .size()
         .rename("homicidios")
    )
    return ts.asfreq("W-SUN", fill_value=0)


ts_total = weekly_series(df)
ts_guayas = weekly_series(df[df["provincia"] == "GUAYAS"])
ts_guayaquil = weekly_series(
    df[df["canton"] == "GUAYQUIL"] if "GUAYQUIL" in df["canton"].unique()
    else df[df["canton"] == "GUAYAQUIL"]
)


def metrics(y_true, y_pred):
    y_true = np.array(y_true, dtype=float)
    y_pred = np.array(y_pred, dtype=float)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    denom = np.where(y_true == 0, np.nan, y_true)
    mape = np.nanmean(np.abs((y_true - y_pred) / denom)) * 100
    return rmse, mae, mape


TRAIN_W = 26


def rolling_arima(ts, train_w=TRAIN_W):
    y = ts.values.astype(float)
    preds, actuals, dates = [], [], []

    for t in range(train_w, len(y)):
        train = y[t - train_w:t]
        z_train = np.log1p(train)

        m = pm.auto_arima(
            z_train,
            seasonal=False,
            stepwise=True,
            suppress_warnings=True,
            error_action="ignore",
            max_p=2, max_q=2, max_d=2,
            information_criterion="aic",
            n_jobs=1
        )
        z_hat = m.predict(n_periods=1)[0]
        y_hat = max(0.0, np.expm1(z_hat))

        preds.append(y_hat)
        actuals.append(y[t])
        dates.append(ts.index[t])

    return pd.Series(preds, index=dates), pd.Series(actuals, index=dates)


def rolling_hw(ts, train_w=TRAIN_W):
    y = ts.values.astype(float)
    preds, actuals, dates = [], [], []

    for t in range(train_w, len(y)):
        train = y[t - train_w:t]
        z_train = np.log1p(train)

        best, best_sse, best_cfg = None, np.inf, None
        for cfg in [
            dict(trend="add", damped_trend=False),
            dict(trend="add", damped_trend=True),
            dict(trend=None, damped_trend=False)
        ]:
            try:
                m = ExponentialSmoothing(
                    z_train, trend=cfg["trend"], seasonal=None, damped_trend=cfg["damped_trend"]
                ).fit(optimized=True)
                if m.sse < best_sse:
                    best, best_sse, best_cfg = m, m.sse, cfg
            except:
                continue

        if best is None:
            best = ExponentialSmoothing(z_train, trend="add", seasonal=None).fit()

        z_hat = float(best.forecast(1)[0])
        y_hat = max(0.0, np.expm1(z_hat))

        preds.append(y_hat)
        actuals.append(y[t])
        dates.append(ts.index[t])

    return pd.Series(preds, index=dates), pd.Series(actuals, index=dates)


def rolling_lstm(ts, lookback=26):
    y = ts.values.astype(float)
    y_log = np.log1p(y)

    scaler = MinMaxScaler()
    y_scaled = scaler.fit_transform(y_log.reshape(-1, 1)).flatten()

    lb = lookback
    X, Y = [], []
    for t in range(lb, len(y_scaled)):
        X.append(y_scaled[t - lb:t])
        Y.append(y_scaled[t])
    X = np.array(X).reshape(-1, lb, 1)
    Y = np.array(Y)

    model = Sequential([LSTM(16, input_shape=(lb, 1)), Dense(1)])
    model.compile(optimizer="adam", loss="mse")
    es = EarlyStopping(monitor="loss", patience=5, restore_best_weights=True)
    model.fit(X, Y, epochs=60, batch_size=16, verbose=0, callbacks=[es])

    preds = model.predict(X, verbose=0).flatten()
    preds_log = scaler.inverse_transform(preds.reshape(-1, 1)).flatten()
    preds_y = np.expm1(preds_log)
    preds_y = np.clip(preds_y, 0, None)

    idx = ts.index[lb:]
    return pd.Series(preds_y, index=idx), pd.Series(y[lb:], index=idx)


def evaluate_rolling(ts, label):
    ts = ts.asfreq("W-SUN", fill_value=0)
    if len(ts) <= TRAIN_W + 5:
        raise ValueError(f"âŒ {label}: serie muy corta ({len(ts)} semanas) para rolling con train=26.")

    print(f"\nðŸ“ {label}")
    print(f"Rango: {ts.index.min().date()} a {ts.index.max().date()} | N={len(ts)} semanas | train_window={TRAIN_W}")

    p_ar, a_ar = rolling_arima(ts, TRAIN_W)
    p_hw, a_hw = rolling_hw(ts, TRAIN_W)
    p_ls, a_ls = rolling_lstm(ts, lookback=TRAIN_W)

    common = p_ar.index.intersection(p_hw.index).intersection(p_ls.index)
    y_true = ts.loc[common].values

    out = []
    out.append(["ARIMA", *metrics(y_true, p_ar.loc[common].values), "auto_arima(max p/q=2) + log1p"])
    out.append(["Holt-Winters", *metrics(y_true, p_hw.loc[common].values), "trend/no-season + log1p"])
    out.append(["LSTM", *metrics(y_true, p_ls.loc[common].values), "1x train, lookback=26 + log1p"])

    df_out = (
        pd.DataFrame(out, columns=["Modelo", "RMSE", "MAE", "MAPE", "Notas"])
          .sort_values("RMSE")
          .reset_index(drop=True)
    )
    display(df_out)

    return df_out, ts.loc[common], p_ar.loc[common], p_hw.loc[common], p_ls.loc[common]


res_total, y_total, p_ar_total, p_hw_total, p_ls_total = evaluate_rolling(ts_total, "TOTAL PAÃS")
res_guayas, y_guayas, p_ar_guayas, p_hw_guayas, p_ls_guayas = evaluate_rolling(ts_guayas, "PROVINCIA: GUAYAS")
res_guayaquil, y_gye, p_ar_gye, p_hw_gye, p_ls_gye = evaluate_rolling(ts_guayaquil, "CANTÃ“N: GUAYAQUIL")

print("\nâœ… Listo. Si quieres, te grafico Real vs Predicho y exporto resultados a Excel.")


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def best_model_series(res_df, preds_dict):
    best_name = res_df.sort_values("RMSE").iloc[0]["Modelo"]
    return best_name, preds_dict[best_name]


def plot_real_vs_preds(y, preds_dict, title):
    plt.figure(figsize=(14, 5))
    plt.plot(y.index, y.values, label="Real")
    for name, series in preds_dict.items():
        plt.plot(series.index, series.values, label=name, alpha=0.85)
    plt.title(title)
    plt.xlabel("Fecha (semanal)")
    plt.ylabel("Homicidios (conteo)")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()


def plot_metric_bars(res_df, title_prefix):
    for metric in ["RMSE", "MAE", "MAPE"]:
        plt.figure(figsize=(8, 4))
        plt.bar(res_df["Modelo"], res_df[metric].values)
        plt.title(f"{title_prefix} - {metric}")
        plt.ylabel(metric)
        plt.xticks(rotation=0)
        plt.grid(True, axis="y", alpha=0.3)
        plt.show()


def plot_residuals(y, pred, title):
    resid = (y - pred).copy()
    plt.figure(figsize=(14, 4))
    plt.plot(resid.index, resid.values)
    plt.axhline(0, linewidth=1)
    plt.title(title)
    plt.xlabel("Fecha (semanal)")
    plt.ylabel("Residual (Real - Predicho)")
    plt.grid(True, alpha=0.3)
    plt.show()


preds_total = {"ARIMA": p_ar_total, "Holt-Winters": p_hw_total, "LSTM": p_ls_total}
preds_guayas = {"ARIMA": p_ar_guayas, "Holt-Winters": p_hw_guayas, "LSTM": p_ls_guayas}
preds_gye = {"ARIMA": p_ar_gye, "Holt-Winters": p_hw_gye, "LSTM": p_ls_gye}


print("ðŸ“Œ TABLA MÃ‰TRICAS - TOTAL PAÃS")
display(res_total)

print("ðŸ“Œ TABLA MÃ‰TRICAS - GUAYAS")
display(res_guayas)

print("ðŸ“Œ TABLA MÃ‰TRICAS - GUAYAQUIL")
display(res_guayaquil)


best_total = res_total.sort_values("RMSE").iloc[0].copy()
best_guayas = res_guayas.sort_values("RMSE").iloc[0].copy()
best_gye = res_guayaquil.sort_values("RMSE").iloc[0].copy()

summary = pd.DataFrame([
    ["TOTAL PAÃS", best_total["Modelo"], best_total["RMSE"], best_total["MAE"], best_total["MAPE"]],
    ["GUAYAS", best_guayas["Modelo"], best_guayas["RMSE"], best_guayas["MAE"], best_guayas["MAPE"]],
    ["GUAYAQUIL", best_gye["Modelo"], best_gye["RMSE"], best_gye["MAE"], best_gye["MAPE"]],
], columns=["Nivel", "Mejor Modelo (RMSE)", "RMSE", "MAE", "MAPE"])

print("ðŸ† RESUMEN: Mejor modelo por nivel (segÃºn RMSE)")
display(summary)


plot_real_vs_preds(y_total, preds_total, "TOTAL PAÃS - Real vs Predicho (Rolling forecast)")
plot_real_vs_preds(y_guayas, preds_guayas, "GUAYAS - Real vs Predicho (Rolling forecast)")
plot_real_vs_preds(y_gye, preds_gye, "GUAYAQUIL - Real vs Predicho (Rolling forecast)")


plot_metric_bars(res_total, "TOTAL PAÃS")
plot_metric_bars(res_guayas, "GUAYAS")
plot_metric_bars(res_guayaquil, "GUAYAQUIL")


best_name_total, best_pred_total = best_model_series(res_total, preds_total)
plot_residuals(y_total, best_pred_total, f"TOTAL PAÃS - Residuales (Mejor: {best_name_total})")

best_name_guayas, best_pred_guayas = best_model_series(res_guayas, preds_guayas)
plot_residuals(y_guayas, best_pred_guayas, f"GUAYAS - Residuales (Mejor: {best_name_guayas})")

best_name_gye, best_pred_gye = best_model_series(res_guayaquil, preds_gye)
plot_residuals(y_gye, best_pred_gye, f"GUAYAQUIL - Residuales (Mejor: {best_name_gye})")


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping


FREQ = "W-SUN"
FORECAST_END = "2030-12-29"
LOOKBACK = 26
SEED = 42

np.random.seed(SEED)
tf.random.set_seed(SEED)
tf.get_logger().setLevel("ERROR")


def best_model_name(res_df):
    return res_df.sort_values("RMSE").iloc[0]["Modelo"]


def forecast_holt_winters(ts, steps):
    y = ts.astype(float).values
    z = np.log1p(y)

    model = ExponentialSmoothing(
        z,
        trend="add",
        seasonal=None,
        initialization_method="estimated"
    ).fit(optimized=True)

    z_hat = model.forecast(steps)
    y_hat = np.expm1(z_hat)
    y_hat = np.clip(y_hat, 0, None)
    return y_hat


def forecast_lstm(ts, steps, lookback=26, epochs=80, batch_size=16):
    y = ts.astype(float).values
    y_log = np.log1p(y).reshape(-1, 1)

    scaler = MinMaxScaler()
    y_scaled = scaler.fit_transform(y_log).flatten()

    if len(y_scaled) <= lookback + 5:
        raise ValueError("Serie demasiado corta para LSTM con lookback indicado.")

    X, Y = [], []
    for i in range(lookback, len(y_scaled)):
        X.append(y_scaled[i - lookback:i])
        Y.append(y_scaled[i])
    X = np.array(X).reshape(-1, lookback, 1).astype(np.float32)
    Y = np.array(Y).astype(np.float32)

    model = Sequential([
        LSTM(16, input_shape=(lookback, 1)),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    es = EarlyStopping(monitor="loss", patience=8, restore_best_weights=True)

    model.fit(X, Y, epochs=epochs, batch_size=batch_size, verbose=0, callbacks=[es])

    history = list(y_scaled)
    preds_scaled = []
    for _ in range(steps):
        x_in = np.array(history[-lookback:], dtype=np.float32).reshape(1, lookback, 1)
        yhat = float(model.predict(x_in, verbose=0)[0, 0])
        preds_scaled.append(yhat)
        history.append(yhat)

    preds_scaled = np.array(preds_scaled).reshape(-1, 1)
    preds_log = scaler.inverse_transform(preds_scaled).flatten()
    preds = np.expm1(preds_log)
    preds = np.clip(preds, 0, None)
    return preds


def future_index(ts, end_date=FORECAST_END, freq=FREQ):
    last = ts.index.max()
    future_idx = pd.date_range(start=last + pd.Timedelta(days=7), end=end_date, freq=freq)
    return future_idx


def forecast_best(ts, res_df, label):
    ts = ts.asfreq(FREQ, fill_value=0)

    best = best_model_name(res_df)
    idx_future = future_index(ts)
    steps = len(idx_future)

    if best.lower().startswith("holt"):
        y_hat = forecast_holt_winters(ts, steps)
    elif best.upper().startswith("LSTM"):
        y_hat = forecast_lstm(ts, steps, lookback=LOOKBACK)
    elif best.upper().startswith("ARIMA"):
        raise ValueError(f"{label}: mejor modelo = ARIMA (no incluido en este bloque).")
    else:
        raise ValueError(f"{label}: modelo no reconocido: {best}")

    fc = pd.Series(y_hat, index=idx_future, name="forecast")
    return best, fc


best_total, fc_total = forecast_best(ts_total, res_total, "TOTAL PAÃS")
best_guayas, fc_guayas = forecast_best(ts_guayas, res_guayas, "GUAYAS")
best_gye, fc_gye = forecast_best(ts_guayaquil, res_guayaquil, "GUAYQUIL")

print("Mejor modelo TOTAL PAÃS:", best_total)
print("Mejor modelo GUAYAS:", best_guayas)
print("Mejor modelo GUAYAQUIL:", best_gye)


def plot_forecast(ts, fc, title):
    plt.figure(figsize=(14, 5))
    plt.plot(ts.index, ts.values, label="Real (histÃ³rico)")
    plt.plot(fc.index, fc.values, label="PronÃ³stico 2030")
    plt.title(title)
    plt.xlabel("Fecha (semanal)")
    plt.ylabel("Homicidios (conteo)")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()


plot_forecast(ts_total, fc_total, f"TOTAL PAÃS | Mejor modelo: {best_total} | ProyecciÃ³n semanal a 2030")
plot_forecast(ts_guayas, fc_guayas, f"GUAYAS | Mejor modelo: {best_guayas} | ProyecciÃ³n semanal a 2030")
plot_forecast(ts_guayaquil, fc_gye, f"GUAYAQUIL | Mejor modelo: {best_gye} | ProyecciÃ³n semanal a 2030")


df_forecasts = pd.concat([
    fc_total.rename("total_pais"),
    fc_guayas.rename("guayas"),
    fc_gye.rename("guayaquil"),
], axis=1)

df_forecasts.to_csv("pronostico_semanal_2026_2030.csv", index=True)
print("âœ… Exportado: pronostico_semanal_2026_2030.csv")
```
